```python
import pandas as pd

# lung.dic and lung.txt are downloaded using SEERStat

features = []
with open('lung_features.txt') as f:
    for line in f:
        features.append(line.split('\n')[0])
features.extend(['Year of diagnosis', 'Patient ID'])

all_features = {}
with open('/home/meerak/seer_ms_real/lung.dic') as f:
    for line in f:
        if 'Var' in line and 'Name' in line:
            all_features[line.split('=')[1].split('\n')[0]] = int(line.split('Var')[1].split('Name')[0]) - 1

final_input_features = []
final_input_features_dict = {}
count = 0
for feature in features:
    final_input_features.append(all_features[feature])
    final_input_features_dict[all_features[feature]] = feature

print(final_input_features)

output_feature = []
for feature in all_features.keys():
    if feature == 'Survival months':
        output_feature = [all_features[feature]]

print(output_feature)

all_final_features = final_input_features.copy()
all_final_features.extend(output_feature)
```

```python
i = -1
column_info = {}
non_categorical_pds = []
new_columns = []
for column in data.columns[:-3]:
    i += 1
    curr_col = list(set(data[column]))
    num_digits = sum([str(j).isdigit() for j in curr_col])
    if num_digits > 1 and num_digits == len(curr_col) - 1 or num_digits == len(curr_col):
        column_info[column] = {}
        print(num_digits, column, curr_col[:2])
        count = 0
        for elem in curr_col:
            if str(elem).isdigit() == False:
                column_info[column][elem] = count
                count += 1
    else:
        non_categorical_pds.append(column)
```

```python
data = data.sort_values(3).groupby(245).first()
```

```python
data = data.reset_index()
```

```python
for c in data.columns:
    print(c)
```

```python
updated_data = data[[245, non_categorical_pds[0], 196]]
updated_data['value'] = [1]*len(updated_data)
updated_data = updated_data.pivot(index = 245, columns = non_categorical_pds[0], values = 'value').fillna(0)
updated_data = updated_data.add_suffix('_%d'%non_categorical_pds[0])
updated_data = updated_data.reset_index()
```

```python
for c in updated_data.columns:
    print(c)
```

```python
for column in non_categorical_pds[1:]:
    curr_updated_data = data[[245, column]]
    curr_updated_data['value'] = [1]*len(curr_updated_data)
    curr_updated_data = curr_updated_data.pivot(index = 245, columns = column, values = 'value').fillna(0)
    curr_updated_data = curr_updated_data.add_suffix('_%d'%column)
    curr_updated_data = curr_updated_data.reset_index()
    updated_data = updated_data.merge(curr_updated_data, on=245)
```

```python
import numpy as np
for key, value in column_info.items():
    curr_updated_data = data[[245, key]].replace(column_info[key])
    updated_data = updated_data.merge(curr_updated_data, on=245)
    updated_data[key] = pd.to_numeric(updated_data[key], errors='coerce')
    updated_data[key] = (updated_data[key] - np.mean(updated_data[key]))/(np.std(updated_data[key]) + 1e-5)
```
